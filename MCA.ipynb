```python
#Petit tuto pour apprendre à utiliser la base de donnée stats_pokemons.db 
#La lecture de la base de donnée se fait via des commandes en SQL et les deux bibliothèques suivantes :
import pandas as pd
import sqlite3
from prince import MCA
import matplotlib.pyplot as plt
conn = sqlite3.connect('STNUM_pokemon.db') #Outil permettant d'exploiter la base de donnée

final_params_number = 100

#On peut ensuite récupérer dans un format dataframe (le même format que pour la STNum) les données via des commandes SQL
request_moves_obtention = 'SELECT * FROM Moves_obtention' 
request_binary_moves_obtention = 'SELECT * FROM binary_description'
df_moves_obtention = pd.read_sql(request_moves_obtention, conn).iloc[:,1:].astype(str)
df_binary_moves_obtention = pd.read_sql(request_binary_moves_obtention, conn).iloc[:,1:].astype(str)
```

```python
import numpy as np
moyenne = pd.DataFrame(pd.read_sql(request_moves_obtention, conn).iloc[:,1:].mean())
moyenne
```

```python
mca = MCA(n_components=final_params_number)

mca.fit(df_moves_obtention)

row_coordinates = mca.row_coordinates(df_moves_obtention)
row_coordinates
```

```python
column_coordinates = mca.column_coordinates(df_moves_obtention)
column_coordinates
```

```python
types_index = {'Grass_Type': [], 'Fire_Type': [], 'Water_Type': [], 'Bug_Type': [], 'Normal_Type': [], 'Poison_Type': [], 'Electric_Type': [], 'Ground_Type': [], 'Fairy_Type': [], 'Fighting_Type': [], 'Psychic_Type': [], 'Rock_Type': [], 'Ghost_Type': [], 'Ice_Type': [], 'Dragon_Type': [], 'Dark_Type': [], 'Steel_Type': [], 'Flying_Type': []}
types_names = ['Grass_Type', 'Fire_Type', 'Water_Type', 'Bug_Type', 'Normal_Type', 'Poison_Type', 'Electric_Type', 'Ground_Type', 'Fairy_Type', 'Fighting_Type', 'Psychic_Type', 'Rock_Type', 'Ghost_Type', 'Ice_Type', 'Dragon_Type', 'Dark_Type', 'Steel_Type', 'Flying_Type']
for name in types_names:
    for i in range(len(df_binary_moves_obtention)):
        if df_binary_moves_obtention.iloc[i][name] == '1':
            types_index[name].append(i)
```

```python
eigenvalue_chosen = 3
data = []
for i in range(len(types_index)):
    pokemon_type_Y = [row_coordinates.iloc[i][eigenvalue_chosen] for i in types_index[types_names[i]]]
    data.append(pokemon_type_Y)
plt.boxplot(data, labels=types_names)
plt.xticks(rotation='vertical')
plt.show()
```

```python
import numpy as np
contribution_matrix_index = column_coordinates.index
X = contribution_matrix_index
Y = column_coordinates[eigenvalue_chosen]
max_indices = np.argsort(Y)[-30:]
X_max = X[max_indices]
Y_max = Y[max_indices]
plt.scatter(X_max, Y_max, color='red')
plt.xticks(rotation='vertical')
plt.show()
```
