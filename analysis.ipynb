```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def dict_to_list(dict):
    list = []
    for key, value in dict.items():
        list.append([key, value])
    return list
```

```python
data_file = 'usage/gen9ou-0.json'
df = pd.read_json(data_file)
df = df.transpose()
df = df.applymap(lambda cell: dict_to_list(cell) if isinstance(cell, dict) else cell)
df = df.loc[:, ['Moves', 'Abilities', 'usage']]
```
